```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
df.describe()
```

```python
plt.scatter(df["YearsExperience"], df["Salary"])
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Years of Experience")
plt.show()
```

```python
X = df[["YearsExperience"]]
Y = df["Salary"]

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42  # 20% of data is used for testing and rest for training
)
```

```python
model = LinearRegression()  # creating empty linear regression model
model.fit(X_train, Y_train) # using the training data to learn the relationship between X and Y
```

```python
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
Y_pred = model.predict(X_test)
```

```python
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
```
